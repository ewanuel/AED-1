```python
#desafio 1

def eh_par(numero):
    if numero % 2 == 0:
        print ("esse número é par")
        return True
    else:
        print ("esse número não é par")
        return False

n = int(input("digite um número n\""))
par = eh_par(n)
print(f" Retorna: {par}")
```

```python
#desafio 2
def calcular_imc(peso, altura):

    if altura <= 0:
        print("Erro: Altura deve ser maior que zero!")
        return None
    if peso <= 0:
        print("Erro: Peso deve ser maior que zero!")
        return None
        
    
    imc = peso / (altura**2)

    return imc

def classificar_imc(imc):

    if imc < 18.5:
        return "Abaixo do peso"
    elif imc < 25:
        return "Peso Normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obseidade"

p = float(input("Favor, Digite o peso"))
a = float(input("Favor, digite a altura:"))

Calculo_IMC = calcular_imc(p, a)

Classifica_IMC = classificar_imc(Calculo_IMC)

print(f"\nSeu IMC é: {Calculo_IMC:.2f}")
print(f"Classificação: {Classifica_IMC}")
```
