```python
#while, repete o bloco enquanto a condição for verdadeira

contador = -5

while contador < 5:
  print(contador)
  contador += 1


lista = ['x', 'x', 'x', 'x', 'x']

while 'x' in lista:
  print(lista)
  lista.pop()

#caso queira deixar mais de um x

lista = ['x', 'x', 'x', 'x', 'x']

while len(lista) > 1: # colocar o numero do indique que deseja parar
  print(lista)
  lista.pop(0)
```

```python
#laço for
for letra in 'python':
  print(letra)

frutas = ["maça", "pera", "abaixi", "uva"]
for fruta in frutas:
    print(fruta)
```

```python
#Função range

#Util para determinar a quantidade de vezes que vamos repetir uma ação

for i in range(10):
  print (i)

print ("================")

#Podemos também definir o espaço da contagem
for e in range(10, 15):
  print (e)

print ("================")

#Podemos também definir o espaço da contagem e o intervalo entre esses números
for e in range(10, 70, 7):
  print (e)
print ("================")

#Também é possivel contar de trás para frente usando o passo negativo
for e in range(70, 10, -7):
  print (e)

```

```python
#Break, interrompe o laço em uma dada condição

Animais = ["lobo", "raposa", "tigre", "leão", "urso"]
for animal in Animais:
  if animal == "leão":
    break
  print(animal)


print ("================")

#Continue, pula o item

for animal in Animais:
  if animal == "leão":
    continue
  print(animal)
```
