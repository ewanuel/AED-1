```python
import random

#pedra>tesoura
#tesoura>papel
#papel>pedra

itens = ["pedra", "papel", "tesoura"]

computador = random.choice(itens)
escolha = input ("selecione um item:\n 1.pedra \n 2.papel \n 3.tesoura\n\n")

if escolha.isdigit():
    indice = int(escolha) - 1
    if 0 <= indice < len(itens):
       escolha_usuario = itens[indice]
       print(f"VocÃª escolheu {escolha_usuario} e o computador escolheu {computador}")

    if escolha_usuario == computador:
      print("empate")
    else:
        if escolha_usuario == "pedra" and computador == "tesoura":
            print("VocÃª ganhou")
        elif escolha_usuario == "papel" and computador == "pedra":
                print("VocÃª ganhou")
        elif escolha_usuario == "tesoura" and computador == "papel":
                print("VocÃª ganhou")
        else:
                print("VocÃª perdeu")

```
