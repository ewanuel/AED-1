```python
num_1 = 60
num_2 = 26

#Operadores Aritmetricos:

print("Operadores Aritmetricos:")
print("Adição: ", num_1 + num_2)
print("Subtração: ", num_1 - num_2)
print("Multiplicação: ", num_1 * num_2)
print("Divisão: ", num_1 / num_2)
print("Divisão Inteira: ", num_1 // num_2)
print("Resto da Divisão: ", num_1 % num_2)
print("Potência: ", num_1 ** num_2)
```

```python
objt_1 = 10
objt_4 = 15

#Operadores de Comparação:

print("Operadores de Comparação:")
print("Igualdade: ", objt_1 == objt_4)
print("Desigualdade: ", objt_1 != objt_4)
print("Maior que: ", objt_1 > objt_4)
print("Menor que: ", objt_1 < objt_4)
print("Maior ou igual que: ", objt_1 >= objt_4)
print ("Menor ou igual que: ", objt_1 <= objt_4)
```

```python
objt_1 = 1.7
objt_2 = 220

#Operadores de Atribuição:

print("Operadores de Atribuição:")
objt_1 += objt_2
print("Adição e Atribuição: ", objt_1)
objt_1 -= objt_2
print("Subtração e Atribuição: ", objt_1)
objt_1 *= objt_2
print("Multiplicação e Atribuição: ", objt_1)
objt_1 /= objt_2
print("Divisão e Atribuição: ", objt_1)
objt_1 %= objt_2
print("Resto da Divisão e Atribuição: ", objt_1)
objt_1 //= objt_2
print("Divisão Inteira e Atribuição: ", objt_1)
objt_1 **= objt_2
print("Potência e Atribuição: ", objt_1)

```

```python
objt_1 = 1.7
objt_2 = 220

#Operadores Logicos:

print("Operadores Logicos:")
print("E: ", objt_1 < objt_2 and objt_1 > objt_2)
print("OU: ", objt_1 < objt_2 or objt_1 > objt_2)
print("Negação: ", not objt_1 < objt_2)
print("Negação: ", not objt_1 > objt_2)
print("Negação com E:" , not (objt_1 < objt_2 and objt_1 > objt_2))
print("Negação com OU: ", not (objt_1 < objt_2 or objt_1 > objt_2))

#
```
