```python
#existem alguns modulos que temos que importar, tal qual o modulo math()

import math

#usando a função sqrt() do módulo math para calcular a raiz quadrada
numero= 16
raiz= math.sqrt(numero)
print(f"A raiz quadrada de {numero} é {raiz}")
```

```python
import matplotlib.pyplot as plt

#Dados para gráfico
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio']
vendas = [1000, 1500, 1200, 1800, 2000]

#criando um grafico de linha
plt.plot(meses, vendas) #cria o grafico com base nos dados fornecidos
plt.title('Vendas Mensais') #adiciona titulo
plt.xlabel('Meses') #adiciona rotulo nos labels
plt.ylabel('Vendas')
plt.grid(True) #coloca a grade
plt.show() #mostra o gráfico
```
