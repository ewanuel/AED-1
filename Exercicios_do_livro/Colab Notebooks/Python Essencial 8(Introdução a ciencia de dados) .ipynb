```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

df =pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

print(df.columns) #verifica colunas extreidas do dataset

df = df.drop(['PassengerId', 'Name', 'Ticket', 'Ticket'], axis=1) #remove dados Axis=1 indica que estamos removendo colunas

print("=============")

print(df.columns)

print("=============")

#preenchendo dados faltantes

df['Age'] = df['Age'].fillna(df['Age'].mean()) #preencheu com a média


sns.histplot(df['Age'], kde=True ,bins=50) #bins é o intervalo entre os valores, quanto maior mais detalhedo kde mostra a curva suave do grafico
plt.title("Distribuição de Idade entre os passageiros")
plt.show
```
