```python

while True:
    tot_num = int(input("Quantos N√∫meros voc√™ deseja inserir na lista?"))
    if tot_num > 0:
        break
    else:
        print("Por favor, digite um n√∫mero maior que zero.")
    
num = []

for i in range(tot_num):
    n = float(input(f"Digite o {i+1}¬∫ n√∫mero: "))
    num.append(n)

maior = max(num)
menor = min(num)
soma = sum(num)
media = soma / len(num)

print(f"\nResultados:")
print(f"Maior n√∫mero: {maior}")
print(f"Menor n√∫mero: {menor}")
print(f"Soma: {soma}")
print(f"M√©dia: {media:.2f}")
```

```python
def jogo_da_forca():
    # Palavra secreta
    palavra_secreta = "PYTHON"
    palavra_secreta = palavra_secreta.upper()
    
    # Vari√°veis do jogo
    letras_descobertas = ['_'] * len(palavra_secreta)
    letras_tentadas = set()
    tentativas_restantes = 6  # Opcional: limitar tentativas
    acertou = False
    
    print("=== JOGO DA FORCA SIMPLIFICADO ===")
    print(f"A palavra tem {len(palavra_secreta)} letras: {' '.join(letras_descobertas)}")
    print(f"Voc√™ tem {tentativas_restantes} tentativas")
    print("-" * 40)
    
    while not acertou and tentativas_restantes > 0:
        # Solicita uma letra do usu√°rio
        letra = input("Digite uma letra: ").upper().strip()
        
        # Valida√ß√£o da entrada
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, digite apenas UMA letra!")
            continue
        
        # Verifica se a letra j√° foi tentada
        if letra in letras_tentadas:
            print(f"Voc√™ j√° tentou a letra '{letra}'. Tente outra!")
            continue
        
        # Adiciona √† lista de letras tentadas
        letras_tentadas.add(letra)
        
        # Verifica se a letra est√° na palavra secreta
        if letra in palavra_secreta:
            print(f"‚úÖ Acertou! A letra '{letra}' est√° na palavra!")
            
            # Revela a letra nas posi√ß√µes corretas
            for i, char in enumerate(palavra_secreta):
                if char == letra:
                    letras_descobertas[i] = letra
            
            # Mostra o progresso
            print(f"Palavra: {' '.join(letras_descobertas)}")
            
            # Verifica se o jogador ganhou
            if '_' not in letras_descobertas:
                acertou = True
                print("\nüéâ PARAB√âNS! Voc√™ acertou a palavra completa!")
                print(f"A palavra era: {palavra_secreta}")
        
        else:
            tentativas_restantes -= 1
            print(f"‚ùå Errou! A letra '{letra}' n√£o est√° na palavra.")
            print(f"Tentativas restantes: {tentativas_restantes}")
            print(f"Palavra: {' '.join(letras_descobertas)}")
        
        # Mostra letras j√° tentadas
        if letras_tentadas:
            print(f"Letras tentadas: {', '.join(sorted(letras_tentadas))}")
        print("-" * 40)
    
    # Mensagem final se o jogador perdeu
    if not acertou:
        print(f"üíÄ Fim de jogo! A palavra era: {palavra_secreta}")

# Vers√£o alternativa com lista de palavras pr√©-definidas
def jogo_da_forca_avancado():
    import random
    
    # Lista de palavras secretas
    palavras = ["PYTHON", "PROGRAMA", "COMPUTADOR", "ALGORITMO", "VARI√ÅVEL", "FUN√á√ÉO"]
    palavra_secreta = random.choice(palavras).upper()
    
    # Resto do c√≥digo igual √† fun√ß√£o anterior
    letras_descobertas = ['_'] * len(palavra_secreta)
    letras_tentadas = set()
    tentativas_restantes = 6
    acertou = False
    
    print("=== JOGO DA FORCA AVAN√áADO ===")
    print(f"A palavra tem {len(palavra_secreta)} letras: {' '.join(letras_descobertas)}")
    print(f"Voc√™ tem {tentativas_restantes} tentativas")
    print("-" * 40)
    
    while not acertou and tentativas_restantes > 0:
        letra = input("Digite uma letra: ").upper().strip()
        
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, digite apenas UMA letra!")
            continue
        
        if letra in letras_tentadas:
            print(f"Voc√™ j√° tentou a letra '{letra}'. Tente outra!")
            continue
        
        letras_tentadas.add(letra)
        
        if letra in palavra_secreta:
            print(f"‚úÖ Acertou! A letra '{letra}' est√° na palavra!")
            
            for i, char in enumerate(palavra_secreta):
                if char == letra:
                    letras_descobertas[i] = letra
            
            print(f"Palavra: {' '.join(letras_descobertas)}")
            
            if '_' not in letras_descobertas:
                acertou = True
                print("\nüéâ PARAB√âNS! Voc√™ acertou a palavra completa!")
        
        else:
            tentativas_restantes -= 1
            print(f"‚ùå Errou! A letra '{letra}' n√£o est√° na palavra.")
            print(f"Tentativas restantes: {tentativas_restantes}")
            print(f"Palavra: {' '.join(letras_descobertas)}")
        
        if letras_tentadas:
            print(f"Letras tentadas: {', '.join(sorted(letras_tentadas))}")
        print("-" * 40)
    
    if not acertou:
        print(f"üíÄ Fim de jogo! A palavra era: {palavra_secreta}")

# Menu principal
def main():
    while True:
        print("\n=== MENU JOGO DA FORCA ===")
        print("1. Jogo Simples (palavra fixa)")
        print("2. Jogo Avan√ßado (palavra aleat√≥ria)")
        print("3. Sair")
        
        opcao = input("Escolha uma op√ß√£o (1-3): ").strip()
        
        if opcao == "1":
            jogo_da_forca()
        elif opcao == "2":
            jogo_da_forca_avancado()
        elif opcao == "3":
            print("Obrigado por jogar! At√© mais!")
            break
        else:
            print("Op√ß√£o inv√°lida! Tente novamente.")

# Executar o programa
if __name__ == "__main__":
    main()
```
