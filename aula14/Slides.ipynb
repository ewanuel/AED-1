```python
#aquecimento
def calcula(op, n1, n2):
    if op == "+":
        r = n1 + n2
    elif op == "-":
        r = n1 - n2
    elif op == "*":
        r = n1 * n2
    elif op == "/":
        r = n1 / n2
    else:
        print("Favor colocar um operador valido")
    return int(r)
    
```

```python
def potencia (base, expoente = 2):
    return base ** expoente
```

```python
print(potencia(3))
print(potencia(3,4))
```

```python
def soma(n1, n2, n3):
    soma = n1+n2+n3
    return soma
```

```python
print(soma(1,2,3))
```

```python
def somar(*args):
    soma = 0
    for i in args:
        soma += i
    return soma
```

```python
print(somar(1,2,3))
```

```python
def central_de_respostas(*args):
    if args:
        qtd = len(args)
        print(qtd, "Mensagem(ns) recebidas(s).")
        i = 1
        for msg in args:
            print(f"Mensagem {i} Recebida: {msg}")
            i = i + 1
    else:
        print("Não tem frases")
```

```python
central_de_respostas("teste", "teste2")
central_de_respostas()
```
