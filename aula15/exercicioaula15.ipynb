```python
def exibir_menu():
    resp_usr = int(input("=========== MENU =========== \n [1]CADASTRAR \n [2]EXIBIR FRASE \n [3]CADASTRAR\n"))
    if resp_usr == 1:
        print("Cadastrando...")
        return True
    elif resp_usr == 2:
        print("Exibindo frase...")
        return True
    elif resp_usr == 3:
        print("Saindo...")
        return False
    else:
        print("Número inválido! Tente novamente.")
    

exibir_menu()
```
